```python
# Step 1: Label Encoding categorical variables
label_encoders = {}
for column in ['Gender', 'Socioeconomic_Status', 'Parental_Education', 'Father_Occupation', 
               'Mother_Occupation', 'Household_Income_Bracket', 'Help_With_Homework_Frequency', 
               'Parent_Teacher_Meetings', 'Parent_Literacy_Level', 'Encouragement_For_University', 
               'Teacher_Qualification', 'School_Infrastructure', 'Extracurricular_Type', 
               'Stress_Levels', 'Test_Anxiety', 'Motivation_Level']:
    le = LabelEncoder()
    student_performance[column] = le.fit_transform(student_performance[column])  # Corrected the indentation
    label_encoders[column] = le  # Store label encoder for future use if needed

# Step 2: Creating interaction features
# Interaction between attendance and continuous assessments
student_performance['Math_Attendance_Interaction'] = student_performance['Math_Continuous_Assessment'] * student_performance['Attendance_Percentage']
student_performance['English_Attendance_Interaction'] = student_performance['English_Continuous_Assessment'] * student_performance['Attendance_Percentage']

# Step 3: Dropping target columns and preparing the target variable
X = student_performance.drop(columns=['Pass_Fail', 'JAMB_Score', 'WAEC_Score'])  # Drop target and other irrelevant columns
y = student_performance['Pass_Fail'].apply(lambda x: 1 if x == 'Pass' else 0)  # Convert Pass/Fail to binary (1 = Pass, 0 = Fail)

# Displaying the first few rows of the engineered dataset
X.head()
```
